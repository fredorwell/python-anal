```python
# Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU
# def spl_cpu(my_str: str):
#     return my_str.split()[0]

df['Cpu_Company'] = df['Cpu'].apply(lambda my_str: my_str.split()[0])
df.head()
```

```python
# Создать новый признак Memory_Amount, 
# который будет содержать только количество Gb памяти без указания типа носителя
def parse_mem(my_str: str):
    my_str = my_str.split()
    res = 0
    for i in my_str:
        if i.endswith('GB') and i[:-2].isdigit():
            res += int(i[:-2])
        elif i.endswith('TB') and i[:-2].isdigit():
            res += int(i[:-2]) * 1024
    return res
        

# list_mem = ['128GB SSD + 1TB HDD', '64GB Flash Storage', '256GB SSD']
# print(f'{parse_mem(list_mem[2])}')
df['Memory_Amount'] = df['Memory'].apply(parse_mem)
df.head()
```

```python
# Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др.)

def mem_type(my_str: str):
    res = []
    if 'HDD' in my_str:
        res.append('HDD')
    if 'SSD' in my_str:
        res.append('SSD')
    if not res:
        return 'other'
    return '/'.join(res)

# list_mem = ['128GB SSD + 1TB HDD', '64GB Flash Storage', '256GB SSD']
# print(f'{mem_type(list_mem[2])}')
df['Memory_Type'] = df['Memory'].apply(mem_type)
df.sample(15)
```

```python
# Удалите признаки Memory и ScreenResolution
df.drop(columns=['Memory', 'ScreenResolution'], inplace=True)
df.sample(15)
```

```python
# 2.1 Создайте признак SSD, который изначально равен 0
# 2.2 Поставьте в признаке SSD 1, если ноутбук действительно с типом носителя SSD


# df['SSD'] = 0
df['SSD'] = df['Memory_Type'].apply(lambda my_str: 1 if 'SSD' in my_str else 0) 
# df['SSD'] = 0
# df.loc['SSD' in df['Memory_Type'], 'SSD'] = 1
df.sample(5)
```

```python
# 2.3 Уберите в признаке Weight значения 'kg' и поменяйте его тип данных на вещественный
df['Weight'] = df['Weight'].str.replace('kg', '')
df['Weight'].astype('float')
df.sample(10)
```

```python
# Создайте датафрейм с клиентами:
# laptop_id - это индексы датафрейма с ноутбуками
# 3.1 Присоедините к таблице clients данные по ноутбукам через метод join
# 3.2 Присоедините к таблице clients данные по ноутбукам через метод merge
clients = pd.DataFrame({
    'client_id': [45, 32, 67, 33, 43],
    'laptop_id': [506, 398, 710, 120, 1999]
})
```

```python
# 4.1 Найдите среднюю стоимость ноутбуков в зависимости от компании производителя
# Отсортируйте от меньшей стоимости к большей

df.groupby('Company').agg({'Price_euros': 'mean'}).sort_values('Price_euros')
```

```python
# 4.2 Найдите минимальную, среднюю и максимальную стоимости ноутбуков в 
# зависимости от производителя процессора

df.groupby('Cpu_Company').agg({'Price_euros': ['mean', 'min', 'max']}).sort_values(('Price_euros', 'mean'))
```

```python
# 4.3 Постройте таблицу с подсчетом количества ноутбуков
# в данных в зависимости от производителя CPU и ОЗУ

df.pivot_table(index='Cpu_Company', columns='Ram', aggfunc='count', values='Product')
```

```python
# 4.4 Постройте таблицу с подсчетом средней стоимости ноутбуков
# в данных в зависимости от операционной системы и GB памяти

df.pivot_table(index='OpSys', columns='Memory_Amount', aggfunc='mean', values='Price_euros', fill_value=0)
```
